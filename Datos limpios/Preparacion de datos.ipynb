**DATOS FACTURAS DE COMPRAS**

```python
path_facturas_compras = r"C:\Users\valer\Desktop\PROYECTO_FINAL_DIPLOMADO\Datos\compras2022-2024.xlsx"
facturas_compras = pd.read_excel(path_facturas_compras)
facturas_compras.head(10)
```

```python
facturas_compras.info()
```

```python
facturas_compras.drop(columns=["Nº", "NUMERO DUI/DIM","CODIGO DE AUTORIZACION","IMPORTE ICE","IMPORTE IEHD","IMPORTE IEHD","IMPORTE IPJ","TASAS",
                               "OTRO NO SUJETO A CREDITO FISCAL","IMPORTES EXENTOS","IMPORTE COMPRAS GRAVADAS A TASA CERO","IMPORTE GIFT CARD",
                               "TIPO COMPRA","CODIGO DE CONTROL","CON DERECHO A CREDITO FISCAL","ESTADO CONSOLIDACION","SUBTOTAL"], errors="ignore", inplace=True)
facturas_compras.columns
```

```python

facturas_compras.rename(columns={
    "NIT PROVEEDOR": "id_proveedor",
    "RAZON SOCIAL PROVEEDOR": "nombre_proveedor",
    "NUMERO FACTURA": "numero_factura",
    "FECHA DE FACTURA/DUI/DIM": "fecha",
    "IMPORTE TOTAL COMPRA": "total_compra",
    "DESCUENTOS/BONIFICACIONES/REBAJAS SUJETAS AL IVA": "descuentos",
    "IMPORTE BASE CF":"importe_base",
    "CREDITO FISCAL":"credito_fiscal"
}, inplace=True)

facturas_compras.head()
```

```python
# Filtrar filas con fechas válidas
facturas_compras = facturas_compras[facturas_compras['fecha'].str.contains(r'\d{1,2}/\d{1,2}/\d{2,4}', na=False)]
```

```python
# Reasignar la columna después de convertir a fecha
facturas_compras = facturas_compras.assign(
    fecha=pd.to_datetime(facturas_compras['fecha'], format='%d/%m/%Y', errors='coerce')
)
```

```python
facturas_compras.info()
```

```python
facturas_compras.head()
```

```python
facturas_compras.info()
```

```python
facturas_compras['total_compra'] = pd.to_numeric(facturas_compras['total_compra'], errors='coerce')
facturas_compras['descuentos'] = pd.to_numeric(facturas_compras['descuentos'], errors='coerce')
facturas_compras['importe_base'] = pd.to_numeric(facturas_compras['importe_base'], errors='coerce')
facturas_compras['credito_fiscal'] = pd.to_numeric(facturas_compras['credito_fiscal'], errors='coerce')
```

```python
# Convertir la columna directamente a flotante
facturas_compras['total_compra'] = facturas_compras['total_compra'].astype(float)
facturas_compras['descuentos'] = facturas_compras['descuentos'].astype(float)
facturas_compras['importe_base'] = facturas_compras['importe_base'].astype(float)
facturas_compras['credito_fiscal'] = facturas_compras['credito_fiscal'].astype(float)
```

```python
facturas_compras['total_compra'] = facturas_compras['total_compra'].fillna(0)
facturas_compras['descuentos'] = facturas_compras['descuentos'].fillna(0)
facturas_compras['importe_base'] = facturas_compras['importe_base'].fillna(0)
facturas_compras['credito_fiscal'] = facturas_compras['credito_fiscal'].fillna(0)
```

```python
facturas_compras.info()
```

```python
facturas_compras.head()
```

```python
facturas_compras.to_csv(r'C:\Users\valer\Desktop\PROYECTO_FINAL_DIPLOMADO\Datos limpios\facturas_compras.csv', index=False)
facturas_compras.to_excel(r'C:\Users\valer\Desktop\PROYECTO_FINAL_DIPLOMADO\Datos limpios\facturas_compras.xlsx', index=False)
```
