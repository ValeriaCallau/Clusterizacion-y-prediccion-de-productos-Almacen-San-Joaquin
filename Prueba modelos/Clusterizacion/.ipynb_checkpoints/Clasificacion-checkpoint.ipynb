```python
import pandas as pd

path_productos = r"C:\Users\valer\Desktop\PROYECTO_FINAL_DIPLOMADO\Datos limpios\productos.xlsx"
path_ventas = r"C:\Users\valer\Desktop\PROYECTO_FINAL_DIPLOMADO\Datos limpios\ventas_productos.xlsx"

productos = pd.read_excel(path_productos)
ventas_productos = pd.read_excel(path_ventas)

productos.head()
```

```python
ventas_productos.head()
```

```python
# Contar valores nulos por columna
nulos_por_columna = productos.isnull().sum()

# Mostrar el resultado
print(nulos_por_columna)
```

```python
ventas_productos.dropna(subset=['fecha'], inplace=True)
```

```python
# Contar valores nulos por columna
nulos_por_columna = ventas_productos.isnull().sum()

# Mostrar el resultado
print(nulos_por_columna)
```

```python
productos.fillna(0, inplace=True)
ventas_productos.fillna(0, inplace=True)
nulos_por_columna = productos.isnull().sum()
print(nulos_por_columna)
print(productos.columns)
print(ventas_productos.columns)
```

```python
# Agrupar las ventas 
ventas_agrupadas = ventas_productos.groupby('id_producto').agg({
    'cantidad': 'sum',  
    'precio_unit': 'mean'     
}).reset_index()

# Renombrar columnas para claridad
ventas_agrupadas.rename(columns={
    'cantidad': 'total_vendido',
    'precio_unit': 'precio_promedio'
}, inplace=True)

# Verificar los datos preparados
ventas_agrupadas.head(10)
```

```python
# Unir los datos de productos con las características agregadas de ventas
datos_combinados = pd.merge(productos, ventas_agrupadas, on='id_producto', how='left')
datos_combinados.fillna(0, inplace=True)
datos_combinados.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Seleccionar las columnas a normalizar
scaler = MinMaxScaler()
datos_combinados[['total_vendido', 'precio_promedio']] = scaler.fit_transform(
    datos_combinados[['total_vendido', 'precio_promedio']]
)

print(datos_combinados[['total_vendido', 'precio_promedio']].head())
```

```python
from sklearn.cluster import KMeans

# Definir el modelo K-Means con un número inicial de clusters
kmeans = KMeans(n_clusters=4, random_state=42)
datos_combinados['cluster'] = kmeans.fit_predict(datos_combinados[['total_vendido', 'precio_promedio']])

# Verificar los clusters asignados
datos_combinados[['id_producto', 'nombre_producto', 'total_vendido', 'precio_promedio', 'cluster']].head()
```

Método de codo 

```python
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(datos_combinados[['total_vendido', 'precio_promedio']])
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graficar los clusters
sns.scatterplot(
    x='total_vendido', 
    y='precio_promedio', 
    hue='cluster', 
    data=datos_combinados, 
    palette='viridis'
)
plt.title('Clusterización de Productos (4 Clusters)')
plt.xlabel('Total Vendido (Normalizado)')
plt.ylabel('Precio Promedio (Normalizado)')
plt.legend(title='Cluster')
plt.show()
```

```python
# Resumen de características por cluster
cluster_summary = datos_combinados.groupby('cluster').agg({
    'total_vendido': 'mean',
    'precio_promedio': 'mean',
    'id_producto': 'count'
}).rename(columns={'id_producto': 'cantidad_productos'})

cluster_summary
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Evaluar silueta para diferentes números de clusters
silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(datos_combinados[['total_vendido', 'precio_promedio']])
    score = silhouette_score(datos_combinados[['total_vendido', 'precio_promedio']], clusters)
    silhouette_scores.append(score)

# Graficar el índice de silueta
plt.plot(range(2, 10), silhouette_scores, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Índice de Silueta')
plt.title('Evaluación de Silueta')
plt.show()
```
